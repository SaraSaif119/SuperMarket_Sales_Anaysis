```python
plt.figure(figsize=(5,5))
new_columns=df.select_dtypes(include=['number'])
sns.heatmap(new_columns.corr(),cmap='YlOrBr',annot=True)
plt.show()
```

```python
df['Product_Type'].value_counts().reset_index()
```

```python
sns.barplot(data=df.head(20),x='Product_Type',y='Total_Price',alpha=0.7, errorbar=None ,width=0.4,color='blue')
plt.xlabel('product name')
plt.ylabel('Price')
plt.title('product name and total price ')
plt.xticks(rotation = 90)
```

```python
# Display the distribution of data based on the quartiles and potential outliers.
plt.figure(figsize=(8, 5))  # Set figure size
ax = sns.boxplot(data=df.tail(30),x='Branch',y='Rating',width=0.4,color='blue')
plt.title('Boxplot of Tow Columns')
plt.xlabel('Branch')
plt.ylabel('Value of Rating')
plt.grid(True)
```

```python
df['Payment'].value_counts().plot(kind='pie', autopct="%0.1f%%", shadow=True)
plt.title('Percentage of each Payment')
plt.ylabel('Payment_Type')
```

```python
Top_city = df['City'].value_counts().index.tolist()

plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='City', order=Top_city )
plt.xticks(rotation=80)
plt.title('Top City')
plt.xlabel('City')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(8, 5))  # Set figure size
ax = sns.barplot(x='Branch',y='gross income',data=df.head(20),capsize=0.2)
plt.title('gross income of each branch')
plt.xlabel('Branch')
plt.ylabel('Values of each column')
plt.grid(True)
```
